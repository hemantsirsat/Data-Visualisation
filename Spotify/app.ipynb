Wordcloud for track name

```python
from wordcloud import WordCloud, STOPWORDS

text = " ".join(str(track) for track in df["track_name"])
word_cloud = WordCloud(collocations=False,background_color="white").generate(text)

plt.figure(figsize=(20,5))
plt.imshow(word_cloud,interpolation="bilinear")
plt.axis("off")
plt.title("WordCloud for track name",fontsize=16,weight="bold", color="#FF0000")

plt.show()
```

```python
df["track_popularity"] = df["track_popularity"].astype("float64")
```

Most popular tracks

```python
most_popular_tracks = df.sort_values(by="track_popularity",ascending=False)
most_popular_tracks.head(5)[["track_name","track_popularity"]]
```

least popular tracks

```python
least_popular_tracks = df.sort_values(by="track_popularity",ascending=True)
least_popular_tracks.head(5)[["track_name","track_popularity"]]
```

Bar graph representing number of songs per song popularity from worst to best

```python
plt.figure(figsize=(20,10))

ax = df["track_popularity"].value_counts().head(20).plot(kind="bar",color=sns.color_palette("tab20"),edgecolor="black",log=True)

for p in ax.patches:
    ax.annotate(int(p.get_height()),(p.get_x()+0.25,p.get_height()+1),ha="center",va="bottom",color="black")
    ax.tick_params("both",labelsize=13)

plt.xlabel("Track popularity[0: least, 100: most]",weight="bold",fontsize=15,color="#FF0000")
plt.ylabel("Number of songs",weight="bold",fontsize=15,color="#FF0000")
plt.title("Distribution of number of songs according to popularity[From least to increasing]")
plt.show()
```

Bar graph representing number of songs per song popularity from best to worst

```python
plt.figure(figsize=(20,10))

ax = df["track_popularity"].value_counts().tail(20).plot(kind="bar",color=sns.color_palette("tab20"),edgecolor="black")

for p in ax.patches:
    ax.annotate(int(p.get_height()),(p.get_x()+0.25,p.get_height()+1),ha="center",va="bottom",color="black")
    ax.tick_params("both",labelsize=13)

plt.xlabel("Track popularity[0: least, 100: most]",weight="bold",fontsize=15,color="#FF0000")
plt.ylabel("Number of songs",weight="bold",fontsize=15,color="#FF0000")
plt.title("Distribution of number of songs according to popularity[From best to decreasing]")
plt.show()
```

Top Genres based on number of tracks

```python
top_genres_on_tracks = df.groupby("playlist_genre").size().reset_index(name="Number of tracks").sort_values(by="Number of tracks",ascending=False)
top_genres_on_tracks
```

```python
plt.figure(figsize=(20,6))

counts = df["playlist_genre"].value_counts()
explode = (0,0,0,0,0,0.1)

counts.plot(kind="pie",colors=colors,explode=explode,autopct="%1.1f%%")

plt.title("Pie chart representing playlist genre",weight="bold",fontsize=15,color="#FF0000")
plt.xlabel("Genre",weight="bold",fontsize=12,color="#FF0000")
plt.ylabel("Percentage",weight="bold",fontsize=12,color="#FF0000")
plt.show()
```

Song having the longest and the shortest duration in minutes

```python
df["duration_ms"] = df["duration_ms"]*((10**-3)/60)
df["duration"] = df["duration_ms"]
df = df.drop("duration_ms",axis=1)
```

```python
duration_sort = df.sort_values(by="duration",ascending=True)
shortest_song = duration_sort.iloc[0]
longest_song = duration_sort.iloc[len(df)-1]
```

```python
print(f"The longest song is `{longest_song.track_name}` with a duration of {longest_song.duration} minutes")
print(f"The shortest song is `{shortest_song.track_name}` with a duration of {shortest_song.duration} minutes")
```

```python
df["track_album_release_date"] = pd.to_datetime(df["track_album_release_date"],format="mixed")
```

Tracks released in the same year

```python
df["year"] = df["track_album_release_date"].dt.year
yearly_tracks = df.groupby(by="year").size().reset_index(name="Number of tracks")
```

```python
plt.figure(figsize=(25,10))

ax = yearly_tracks["Number of tracks"].plot(kind="bar",edgecolor="black",log=True,color=sns.color_palette("tab20"))
ax.set_xticklabels(yearly_tracks["year"])

for p in ax.patches:
    ax.annotate(int(p.get_height()),(p.get_x()+0.25,p.get_height()+1),va="bottom",ha="center",color="black")
    ax.tick_params("both",labelsize=15)

plt.xlabel("Year",color="#FF0000",weight="bold",fontsize=14)
plt.ylabel("Number of tracks",color="#FF0000",weight="bold",fontsize=14)
plt.title("Number of tracks in a year",color="#FF0000",weight="bold",fontsize=15)
plt.show()
```
